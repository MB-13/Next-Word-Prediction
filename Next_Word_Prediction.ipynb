```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
with open('/content/dialogs.txt','r') as file_:
  data= file_.read()
```

```python
len(data)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
no_of_tokens = len(tokenizer.word_index)
no_of_tokens
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for line in data.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([line])[0]

  if len(tokenized_sentence) > 2:
    for i in range(1,len(tokenized_sentence)):
      input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_sequence = max([len(x) for x in input_sequences])
max_sequence
```

```python
from keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen=max_sequence,padding='pre')
padded_input_sequences[:5]
```

```python
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
print(x.shape)
print(y.shape)
print(max(y))
```

```python
from keras.utils import to_categorical
y = to_categorical(y,num_classes = no_of_tokens+1)
```

```python
y.shape
```

```python
# CREATING MODEL

model = keras.Sequential()

model.add(keras.layers.Embedding(no_of_tokens+1,100,input_length=max_sequence-1))
model.add(keras.layers.Bidirectional(keras.layers.LSTM(1000,return_sequences=True)))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Bidirectional(keras.layers.LSTM(500)))

model.add(keras.layers.Dense(512,activation='relu'))
model.add(keras.layers.Dense(no_of_tokens+1,activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=50)
```

```python
text = " I want"
for i in range(3):
  # tokenizer
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text],maxlen=max_sequence-1,padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
```
